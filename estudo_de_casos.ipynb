# Estudo de Casos

Objetivo: Analisar o comportamento da função objetivo dada e seu domínio.

## Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt

from objective_function_util import f
```

## Estudo

Como a função objetivo f é uma função $$f : \mathbb{R}^5 \Rightarrow \mathbb{R},$$ temos que não é possível plotar ela em um gráfico de duas dimensões. Contudo, ainda é possível estudar a geometria da função fixando o valor de todas as suas variáveis com exceção de uma, e variando esta última variável, obtendo um tipo de seção da hiper-superfície produzida pela função. Logo, faremos alguns gráficos onde o valor de quase todas as variáveis está fixo e uma delas está variando.

### Gráfico 1

Fixando o valor das variáveis x_2 à x_5 com o valor 5, temos que o argumento do logaritmo de f $$\prod_{i=1}^{5}x_i(1-x_i)+1 = x_1(1-x_1)(5(1-5))^4+1,$$ é maior ou igual a 1 se $$x_1(1-x_1)20^4 \geq 0,$$ uma opção de domínio para x_1 que satisfaz esta inequação é (0, 1). Vejamos a seguir como é o gráfico da função para este domínio.

```python
x_1 = np.linspace(0, 1)
```

```python
var_1 = [[i, 5, 5, 5, 5] for i in x_1]
```

```python
y_1 = []

for arg in var_1:
    prod = 1
    for val in arg:
        prod = prod * val * (1 - val)
    if prod < 0:
        print(f"Invalid arg: {arg, prod}")
    else:
        y_1.append(f(arg))

plt.plot(x_1, y_1)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

Podemos observar que a função possui uma taxa de crescimento e decrescimento muito grande próxima dos valores x_1 = 0 e x_1 = 1, respectivamente. E que no meio do domínio a taxa de variação da função é pequena. Como a função neste domínio é concâva para baixo, temos que seus pontos de mínimo para este domínio estariam próximos das extremidades do domínio, isto é, x_1 = 0 e x_1 = 1.

### Gráfico 2

Fixando o valor das variáveis x_2 à x_5 com os valores -5, -3, -2, e 0.5, temos que o argumento do logaritmo de f $$\prod_{i=1}^{5}x_i(1-x_i)+1 = x_1(1-x_1)(-540)+1,$$ é maior ou igual a 1 se $$x_1(1-x_1)(-540) \geq 0,$$ uma opção de domínio para x_1 que satisfaz esta inequação é (-1, 0). Vejamos a seguir como é o gráfico da função para este domínio.

```python
x_2 = np.linspace(-1, 0)
```

```python
var_2 = [[i, -5, -3, -2, 0.5] for i in x_2]
```

```python
y_2 = []

for arg in var_2:
    prod = 1
    for val in arg:
        prod = prod * val * (1 - val)
    if prod < 0:
        print(f"Invalid arg: {arg, prod}")
    else:
        y_2.append(f(arg))

plt.plot(x_2, y_2)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

Podemos observar que a função é decrescente e possui mínimo próximo de x_1 = 0.
